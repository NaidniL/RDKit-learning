数据来自审计批次 `20260704_100408_933902_UTC_07d4f39d`。共 11 个代码单元，每个单元单排绘制 4 个分子。图例中 `INCLUDE` 和 `EXCLUDE` 为用户确认的最终人工决策。

```python
from rdkit import Chem
from rdkit.Chem import Draw

def draw_review_batch(items):
    mols = []
    legends = []
    for number, recommendation, inchikey, smiles, name in items:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            raise ValueError(f'No. {number} SMILES cannot be parsed: {smiles}')
        mols.append(mol)
        legends.append(f'{number}. {name}\n{recommendation} | {inchikey[:14]}')
    return Draw.MolsToGridImage(
        mols, molsPerRow=4, subImgSize=(320, 260), legends=legends, useSVG=False
    )

batch = [
    (1, 'EXCLUDE', 'ATDGTVJJHBUTRL-UHFFFAOYSA-N', 'N#CBr', 'Cyanogen bromide'),
    (2, 'INCLUDE', 'BABMCXWQNSQAOC-UHFFFAOYSA-M', 'C[Hg]Cl', 'Methylmercury chloride'),
    (3, 'INCLUDE', 'BAVYZALUXZFZLV-UHFFFAOYSA-N', 'CN', 'Methylamine / methylamine hydrochloride'),
    (4, 'INCLUDE', 'BDAGIHXWWSANSR-UHFFFAOYSA-N', 'O=CO', 'Formic acid / sodium formate'),
]
draw_review_batch(batch)
```
